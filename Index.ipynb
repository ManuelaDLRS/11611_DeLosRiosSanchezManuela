```python
"""Ejercicio 8"""
def leer_lista_numeros():

    while True:
        try:
            lista= list(eval(input("Ingerese la lista de números separando cada \
                                  termino con comas ")))
            for x in range(len(lista)-1):
                if type(lista[x]) != int:
                    raise ValueError
            break
        except:
            print("Oops... Los números ingresados a la lista no están ordenados\
                   ascendentemente")
    return(lista)
   

lista = leer_lista_numeros()
numero_mayor_1 = -float("inf")
numero_mayor_2 = -float("inf")
for x in lista:
    if x > numero_mayor_1:
        numero_mayor_1 = x
lista.remove(numero_mayor_1)
for x in lista:
    if x > numero_mayor_2:
        numero_mayor_2 = x
print (f"El segundo número mayor de la lista es {numero_mayor_2}")
```

```python
"""Ejercicio 10"""
def leer_matriz():
    """Función que lee número de filas, columnas y elementos de la matriz.
    Si los elementos no son enteros devuelve error.
    Si la cantidad de elementos no coincide con la dimensión devuelve error.
    """
    while True:
        try:
   
            fila = int(input("Ingrese el número de filas de la matriz "))
            col = int(input("Ingrese el número de columnas de la matriz "))
            if fila <= 0 :
                raise ValueError
            if fila//1 != fila:
                raise ValueError
            if col <= 0 :
                raise ValueError
            if col//1 != fila:
                raise ValueError
        except:
            print(f"La dimensión de la matriz no es válida, intente de nuevo")
        try:
            lista_elem_mat = list(eval(input("Ingrese los elementos de la matriz \
en el sentido de las filas ")))
            for elemento in lista_elem_mat:
                if type(elemento) != int:
                    raise ValueError
            if len(lista_elem_mat) != (fila*col):
                raise ValueError
        except:
            print("Los elementos de la matriz no son válidos, intente de nuevo")
    return(fila,col,lista_elem_mat)
   
def crear_mat():
    """Función que retorna una matriz.
    Con la funcion leer_matriz extrae la cantidad de filas, columnas y la \
    lista de elementos que conforman la matriz
    """
    fila,col,lista = leer_matriz()
    contador = 0
    matriz = []
    for i in range (fila):                     #Se crea una matriz de Nones de        
        matriz.append([0]*col)                 #la dimensión requerida
   
    for i in range (fila):                     #A través de un ciclo for se
        for j in range (col):                  #introducen los datos de forma
            matriz[i][j] = lista[contador]     #matricial
            contador += 1
    return(matriz)

def rot_90(matriz):
    lista_new = []
    fila,col,lista = leer_matriz()
   
    for j in range (1,col+1):                   #Leer la lista desde la última
        indice = col-j                          #columna hasta la primera
        for i in range (fila):                  #para generar la rotación de 90°
            lista_new.append(lista[i][indice])
    return(lista_new)
   
new_list = []
matriz = crear_mat()
   
rot_90(matriz, 2)
new_list = new_list + [matriz.pop(0)]

while matriz != 0:
    rot_90(matriz)
    new_list = new_list + (matriz[0])
   
print(new_list)
```

```python
"""Ejercicio 13"""
def num_bis(b):
    'Define la función que identifica si el año es bisiesto'
    return b % 400 == 0 or (b % 100 != 0 and b % 4 == 0)

print('Contador de años')
primer_año = int(input('Escriba un año: ')) #primer año del contador
print('Escriba otro año posterior a', str(primer_año) + ': ')
segun_año = int(input()) #siguiente año del contador
while segun_año < primer_año: #primer año mayor que el segundo
    print(segun_año,' no es mayor que ',str(primer_año),'. Inténtelo de nuevo:')
    segun_año = int(input())

año_bisiestos = 0 #años bisiestos
for x in range(primer_año, segun_año + 1):
    if num_bis(x):
        año_bisiestos += 1 #sumar 1 a la función si esta da año bisiesto
    
print('De ',primer_año,' a ',segun_año,' hay ',segun_año - primer_año + 1,
      ' años, ',año_bisiestos, 'de ellos bisiestos.')
```
